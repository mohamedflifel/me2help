```python
# Load EDA Pkgs
import pandas as pd
import numpy as np
```

```python
# Load Data Viz Pkgs
import seaborn as sns
```

```python
# Load Text Cleaning Pkgs
import neattext.functions as nfx

```

```python
# Load ML Pkgs
# Estimators
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB

# Transformers
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
# Load Dataset
df = pd.read_csv("data/emotion_dataset_raw.csv")
```

```python
df.head()
```

```python
# Value Counts
df['Emotion'].value_counts()
```

```python
# Plot
sns.countplot(x='Emotion',data=df)
```

```python
# Data Cleaning
dir(nfx)
```

```python
# User handles
df['Clean_Text'] = df['Text'].apply(nfx.remove_userhandles)
```

```python
# Stopwords
df['Clean_Text'] = df['Clean_Text'].apply(nfx.remove_stopwords)
```

```python
df
```

```python
# Features & Labels
Xfeatures = df['Clean_Text']
ylabels = df['Emotion']
```

```python
#  Split Data
x_train,x_test,y_train,y_test = train_test_split(Xfeatures,ylabels,test_size=0.3,random_state=42)
```

```python
# Build Pipeline
from sklearn.pipeline import Pipeline
```

```python
# LogisticRegression Pipeline
pipe_lr = Pipeline(steps=[('cv',CountVectorizer()),('lr',LogisticRegression())])
```

```python
# Train and Fit Data
pipe_lr.fit(x_train,y_train)
```

```python
pipe_lr
```

```python
# Check Accuracy
pipe_lr.score(x_test,y_test)
```

```python
ex1="Today felt strangely quiet, but not in a comforting way. I kept checking the door without really knowing why, and every unexpected sound pulled my attention away from what I was doing. I tried to focus on normal things — reading, listening to music — yet my thoughts kept drifting back to small “what if” questions I couldn’t answer. Nothing actually happened, and everything looked the same as always, but I couldn’t fully relax. By the end of the day, I was tired from thinking too much and hoping tomorrow would feel more certain and steady."
```

```python
pipe_lr.predict([ex1])
```
